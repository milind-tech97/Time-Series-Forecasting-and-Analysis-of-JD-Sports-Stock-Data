# Time Series Forecasting Methods

## Naive Forecast Method

```python
train_len = 48# We consider the first 48 observation as traing set which constitutes to 80% of the data.
train = df[0:train_len]
test = df[train_len:] # The last 12 years are considered as out-of-time test set which constitutes to 20% of the d.
```

```python
# Forecast using the Naive Method:
# Inputs:
# Data,list of Historical data points reperesenting the time series.
# nforecast, number of forecast that we want to make into the future.
def naive_method(data, nforecast):

    forecast = [data[-1]]  # The initial forecast is the last observed point
    num = nforecast-1 # We can adjust the number of forecast that we want to make

    # To forecast the next 12 points, we need a loop to iterate:
    for i in range(num):
         forecast.append(forecast[-i])

    return forecast #  returns forecast (list): A list of forecasts for each period.
```

```python
# Here we are creating a new DataFrame y_hat_naive and copying the test DataFrame.
y_hat_naive = test.copy()
# Creating a new Column named 'naive_forecast' with the naive method we are forecasting the future values based on the last observed value in the training set.
# nforecast generates a forecast of next 12 points using the Naive method
y_hat_naive['naive_forecast'] = naive_method(train['Close Price'], nforecast = 12)
```

```python
y_hat_naive
```

```python
# Plotting the Visualization of Train Test & Naive forecast values.
plt.figure(figsize=(13,6))
plt.grid(color='black',alpha=0.2) # setting grid color as black
plt.plot(train['Close Price'], label='Training Data',color='blue')
plt.plot(test['Close Price'], label='Testing Data',color='green')
plt.plot(y_hat_naive['naive_forecast'], label='Naive Method Forecast',color='red')
plt.legend(loc='best',frameon=True) # Adding box arround the legend
plt.title('Naive Forecast',fontdict={'fontsize':12,'fontstyle':'oblique','fontweight':'bold','color':'Black'}) # Adding additional styles to the title
plt.show()
```

### Accurcary metrics

```python
# Here we are caluculating:
# Root Mean Squared Error(RMSE),Mean Absolute Percentage Error(MAPSE) & Mean Absolute Error(MAE) for the forecast generated by Naive Method.

# Importing Linbraries
from sklearn.metrics import mean_squared_error
#Calculates the Root Mean Squared Error between the actual and the forecasted values.
rmse = np.sqrt(mean_squared_error(test['Close Price'], y_hat_naive['naive_forecast'])).round(2)
# Calculates the Mean Absolute Percentage Error between the actual and the forecasted values.
mape = np.round(np.mean(np.abs(test['Close Price']-y_hat_naive['naive_forecast'])/test['Close Price'])*100,2)
# Calculates the Mean Absolute Error between the actual and the forecasted values.
mae  = np.round(np.mean(np.abs(test['Close Price']-y_hat_naive['naive_forecast'])),2)

# Creating the DataFrame named results with Columns Naive method,MAPE,RMSE & MAE.
# Loading the DataFrame with the calculated metrics of Naive Method
results = pd.DataFrame({'Method':['Naive method'], 'MAPE': [mape], 'RMSE': [rmse], 'MAE': [mae]})
results = results[['Method', 'RMSE', 'MAE', 'MAPE']]
results
```

## Average Historical Method

```python
# Average Historical Method
# Inputs:
# Data,list of Historical data points reperesenting the time series.
# nforecast, number of forecast that we want to make into the future.

def average_historical_method(data, nforecast):

   # Calculating Average,the average is calculated by summing all the values divided by the number of observation.
    average = sum(data) / len(data)
    # Calculating Forecast,Creating list named forecast which consists nforecast copies of the calculated average value.
    forecast = [average] * nforecast  # Here we are creating a list for forcaste with same average value.

    return forecast # returns list od forecast of each period
```

```python
# Creating a new DataFrame y_hat_average_hist by copying the test DataFrame
y_hat_average_hist = test.copy()
# Forecasting the values for next 12 points using the Average Historical Method
y_hat_average_hist['average_hist_forecast'] = average_historical_method(train['Close Price'], nforecast = 12)
```

```python
# rounding all the values in the DataFrame y_hat_average_hist to two decimal places.
round(y_hat_average_hist,2)
```

```python
# Ploting the Visualization of Train,Test & Average Historical Method.
plt.figure(figsize=(13,6))
plt.grid(color='black',alpha=0.2)
plt.plot(train['Close Price'], label='Training Data',color='blue')
plt.plot(test['Close Price'], label='Testing Data',color='green')
plt.plot(y_hat_average_hist['average_hist_forecast'], label='Average Historical Method Forecast',color='red')
plt.legend(loc='best',frameon=True)
plt.title('Average Historical Method',fontdict={'fontsize':12,'fontstyle':'oblique','fontweight':'bold','color':'Black'})
plt.show()
```

### Accuracy metrics

```python
# Accuracy metrics
# Calculating the Root Mean Squared Error(RMSE), Mean Absolute Percentage Error(MAPSE), and Mean Absolute Error(MAE) for the forecast generated by Average Historical method.
from sklearn.metrics import mean_squared_error

# Calculates the Root Mean Squared Error between the actual and the forecasted values.
rmse = np.sqrt(mean_squared_error(test['Close Price'], y_hat_average_hist['average_hist_forecast'])).round(2)

# Calculates the Mean Absolute Percentage Error between the actual and the forecasted values.
mape = np.round(np.mean(np.abs(test['Close Price'] - y_hat_average_hist['average_hist_forecast']) / test['Close Price']) * 100, 2)

# Calculates the Mean Absolute Error between the actual and the forecasted values.
mae = np.round(np.mean(np.abs(test['Close Price'] - y_hat_average_hist['average_hist_forecast'])), 2)

# Creating a DataFrame named results with columns Method, MAPE, RMSE & MAE
# Loading the DataFrame with the calculated metrics of Average Historical Method
results = pd.DataFrame({'Method': ['Average all historical Method'], 'MAPE': [mape], 'RMSE': [rmse], 'MAE': [mae]})
results = results[['Method', 'RMSE', 'MAE', 'MAPE']]
results
```

## Exponential Smoothing methods

The Exponetial Smoothing Method uses the Weighted average of past time series value as a forecast(Camm et al., 2024).
There are 3 different Exponential Smoothing Method:
1. Simple/Single Exponential Smoothing Method
2. Holt's Linear Trend Method
3. Hotl-Winters Method

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt

# The Smoothing factor alpha is set 0.6
# Which controls the weight assigned to most recent observation.
alpha = 0.6

y_hat_avg = test.copy()

# Function SimpleExpSmoothing represents the method which is used to fit Simple Exponential Smoothing Model to traing data.
#fit2 = SimpleExpSmoothing(np.asarray(train['Close Price'])).fit(smoothing_level=alpha, optimized=False)
#The Smoothing factor is automatically optimized
fit2 = SimpleExpSmoothing(np.asarray(train['Close Price'])).fit()
#The fitted model is used to forecast the values for the test set.
y_hat_avg['SES'] = fit2.forecast(len(test)) # number of forecasting equal to the test size

# Plotting the Visualization
plt.figure(figsize=(10,6))
plt.grid(color='black',alpha=0.4)
plt.plot(train['Close Price'], label='Training Data',color='blue',linestyle='--')
plt.plot(test['Close Price'], label='Testing Data',color='orange',marker='o')
plt.plot(y_hat_avg['SES'], label='Simple Exponential smothing',color='red')
plt.legend(loc='best',frameon=True)
plt.title("Simple Exponential Smoothing Method",fontdict={'fontsize':15,'fontstyle':'italic','fontweight':'bold','color':'Black'})
plt.show()
```

## Agumented Dicky-Fuller (ADF Test)

```python
# Conducting an Agumented Dicky-Fuller(ADF) Test on the time series Data.
from statsmodels.tsa.stattools import adfuller
from numpy import log
# Applying ADF on the Time Series Data
# dropna() is used to drop the missing values from the "Close" column
result = adfuller(df.Close_Price.dropna())
# Printing the ADF statistics
print('ADF Statistic: %f' % result[0])
# Printing the p-value associated with ADF Test
print('p-value: %f' % result[1])
```

```python
# Importing Libraries and function for plotting Autocorrelation & Partial-Autocorrelation
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})


plot_color = 'blue'

# Plotting the Original Time-Series
fig, axes = plt.subplots(3, 2, sharex='col')
axes[0, 0].plot(df.Close_Price,color='blue'); axes[0, 0].set_title('Original Series')
plot_acf(df.Close_Price, lags = 20, ax=axes[0, 1],color=plot_color)

# Plotting the autocorrelation with 1st Differencing
axes[1, 0].plot(df.Close_Price.diff(),color='blue'); axes[1, 0].set_title('1st Order Differencing')
plot_acf(df.Close_Price.diff().dropna(), lags = 20, ax=axes[1, 1],color=plot_color)

# Plotting the autocorrelation with 2nd Differencing
axes[2, 0].plot(df.Close_Price.diff().diff(),color='blue'); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(df.Close_Price.diff().diff().dropna(), lags = 20, ax=axes[2, 1],color=plot_color)

plt.show()
```

```python
# PACF to identify the AR(p)terms suitable for time series.
# PACF plot of 1st differenced series
plot_pacf(df.Close_Price.diff().dropna(), lags=10)
plt.title('Partial Autocorrelation',fontdict={'fontsize':12,'fontstyle':'oblique','fontweight':'bold','color':'Black'})
plt.show()
```

## Accuracy Metrics for Time Series Forecast

```python
# Accuracy metrics
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    corr = np.corrcoef(forecast, actual)[0,1]   # corr
    mins = np.amin(np.hstack([forecast[:,None],
                              actual[:,None]]), axis=1)
    maxs = np.amax(np.hstack([forecast[:,None],
                              actual[:,None]]), axis=1)
    minmax = 1 - np.mean(mins/maxs)             # minmax
    acf1 = acf(fc-test)[1]                      # ACF1
    return({'mape':mape, 'me':me, 'mae': mae,
            'mpe': mpe, 'rmse':rmse, 'acf1':acf1,
            'corr':corr, 'minmax':minmax})

forecast_accuracy(fc.values, test.values)
```
